```python
import win32com.client

# SAP GUI Scripting 객체 가져오기
SapGuiAuto = win32com.client.GetObject("SAPGUI")
application = SapGuiAuto.GetScriptingEngine

# SAP 연결 설정
connection = application.Children(0)
session = connection.Children(0)

# 헤더까지
session.findById("wnd[0]").maximize()
session.findById("wnd[0]/tbar[0]/okcd").text = "fbb1"
session.findById("wnd[0]").sendVKey(0)
session.findById("wnd[0]/usr/ctxtBKPF-BLDAT").text = "2024.10.31" #증빙일
session.findById("wnd[0]/usr/ctxtBKPF-BLART").text = "C4"
session.findById("wnd[0]/usr/ctxtBKPF-WAERS").text = "eur"
session.findById("wnd[0]/usr/ctxtBKPF-BUKRS").text = "DE70" 
session.findById("wnd[0]/usr/ctxtBKPF-BUDAT").text = "2024.10.31" #전기일
session.findById("wnd[0]/usr/txtBKPF-MONAT").text = "10" #전기기간





import xlwings as xw
import pandas as pd

# 엑셀 파일 열기
wb = xw.Book('w.xlsx')  # 엑셀 템플릿 파일 경로
sheet = wb.sheets['Sheet1']  # 데이터가 있는 시트 이름

# 데이터를 DataFrame으로 읽어오기
df = sheet.range('A1').options(pd.DataFrame, expand='table', header=1, index=False).value

# DataFrame을 반복문으로 처리
# DataFrame을 반복문으로 처리
for index, row in df.iterrows():
    차대 = row['차대']
    통화 = row['통화']
    계정명 = str(int(row['계정명']))
    거래금액 = row['거래금액']
    현지금액 = row['현지금액']
    프로젝트코드 = row['프로젝트코드']
    적요 = row['적요']
    WBS = row['WBS']
    지정 = row['지정']
    원화 = row['원화']
    전역회사통화 = row['전역회사통화']
    거래처 = str(int(row['거래처']))

    # 자동화할 코드
    session.findById("wnd[0]/usr/ctxtRF05A-NEWBS").text = 차대
    session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").text = str(계정명)
    session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").setFocus()
    session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").caretPosition = 8
    session.findById("wnd[0]").sendVKey(0)
    session.findById("wnd[0]/usr/txtBSEG-WRBTR").text = str(거래금액)
    session.findById("wnd[0]/usr/txtBSEG-DMBTR").text = str(현지금액)
    session.findById("wnd[0]/usr/ctxtBSEG-MWSKZ").text = "V0"
    session.findById("wnd[0]/usr/txtBSEG-ZUONR").text = str(프로젝트코드)
    session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").text = str(적요)

    # 계정명이 1이나 2로 시작하지 않으면 WBS 입력
    if not (계정명.startswith('1') or 계정명.startswith('2')):
        session.findById("wnd[0]/usr/subBLOCK:SAPLKACB:9002/ctxtCOBL-PS_POSID").text = str(WBS)

    session.findById("wnd[0]/usr/subBLOCK:SAPLKACB:9002/ctxtCOBL-PRCTR").text = str(지정)
    session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").setFocus()
    session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").caretPosition = len(str(적요))
    session.findById("wnd[0]").sendVKey(7)
    session.findById("wnd[0]/usr/txtBSEG-DMBE2").text = 원화
    session.findById("wnd[0]/usr/txtBSEG-DMBE3").text = 전역회사통화
    session.findById("wnd[0]/usr/ctxtBSEG-XREF1").text = str(거래처)
    session.findById("wnd[0]/usr/ctxtBSEG-XREF2").text = "B"
    session.findById("wnd[0]").sendVKey(14)

    

# 엑셀 파일 닫기
# wb.close()





# session.findById("wnd[0]/usr/ctxtRF05A-NEWBS").text = "40" #차대
# session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").text = "46040110" #계정명
# session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").setFocus()
# session.findById("wnd[0]/usr/ctxtRF05A-NEWKO").caretPosition = 8
# session.findById("wnd[0]").sendVKey(0)
# session.findById("wnd[0]/usr/txtBSEG-WRBTR").text = "1575.21" #거래금액
# session.findById("wnd[0]/usr/txtBSEG-DMBTR").text = "1575.21" #현지금액
# session.findById("wnd[0]/usr/ctxtBSEG-MWSKZ").text = "V0" 
# session.findById("wnd[0]/usr/txtBSEG-ZUONR").text = "TDGDE1801" #프로젝트코드
# session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").text = "프로젝트손익결산 7" #적요
# session.findById("wnd[0]/usr/subBLOCK:SAPLKACB:9002/ctxtCOBL-PS_POSID").text = "TDGDE1801" #WBS
# session.findById("wnd[0]/usr/subBLOCK:SAPLKACB:9002/ctxtCOBL-PRCTR").text = "TDGDE1801" #지정
# session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").setFocus()
# session.findById("wnd[0]/usr/ctxtBSEG-SGTXT").caretPosition = 26
# session.findById("wnd[0]").sendVKey(7)
# session.findById("wnd[0]/usr/txtBSEG-DMBE2").text = "2358893"  #원화
# session.findById("wnd[0]/usr/ctxtBSEG-XREF1").text = "3098243" #거래처
# session.findById("wnd[0]/usr/ctxtBSEG-XREF2").text = "B" 
# session.findById("wnd[0]").sendVKey (14)



```
